## udemy paid or free?

```python
import numpy as np
import pandas as pd
import seaborn as sns
sns.set_style('whitegrid')
import matplotlib.pyplot as plt
%matplotlib inline
import datetime as dt
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv("DATASET/udemy_courses.csv")
df.head(1)
```

```python
df.describe()
```

```python
def converthours(val):
    values=val.split()
#     print(values)
    if values[0]=='0':
        return 0
    elif values[1]=='hours' or values[1]=='hour':
        return float(values[0])
    elif values[1]=='mins':
        return float(values[0])/60

df['content_duration']=df['content_duration'].apply(converthours)
df.isna().any()
df[df['content_duration'].isna()]=0
```

```python
df.head(1)
df[df['published_timestamp']==0]='2012'
```

```python
def getyear(val):
    return val[0:4]
df['year']=df['published_timestamp'].apply(getyear)
df.head(1)
```

```python
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.metrics import classification_report,confusion_matrix
```

```python
df.drop(['course_id','course_title','url','published_timestamp','clean_course_title','price'],axis=1,inplace=True)
df.head(1)
```

```python
def conversion(val):
    if val=='TRUE':
        return 1
    else:
        return 0

df['is_paid']=df['is_paid'].apply(conversion)
```

```python
df.head(1)
```

```python
def lev(val): 
  
    if val == 'All Levels': 
        return 0
    elif val == 'Intermediate Level': 
        return 1
    elif val == 'Beginner Level': 
        return 2
    else:
        return 3
    
df['level'] = df['level'].apply(lev)
```

```python
def sub(val): 
  
    if val == 'Business Finance': 
        return 0
    elif val == 'Graphic Design': 
        return 1
    elif val == 'Musical Instruments': 
        return 2
    else:
        return 3
    
df['subject'] = df['subject'].apply(sub)
```

```python
df.info()
```

```python
def year(val): 
#     print(val==2017)
#     print(type(val),val,type('2017'))
    if val == '2017': 
        return 6
    elif val == '2016': 
        return 5
    elif val == '2015': 
        return 4
    elif val == '2014': 
        return 3
    elif val == '2013': 
        return 2
    elif val == '2012': 
        return 1
    else:
        return 0
    
df['year'] = df['year'].apply(year)
```

```python
df
```

```python
x=df.drop(['is_paid'],axis=1)
x
```

```python
y=df['is_paid']
y
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.30, random_state=101)
```

## logistic regression

```python
from sklearn.linear_model import LogisticRegression
lmodel = LogisticRegression()
lmodel.fit(x_train,y_train)
lpredictions = lmodel.predict(x_test)
print(lpredictions,y_test)
print("Confusion Matrix")
print(confusion_matrix(y_test,lpredictions))

lscore = round((lmodel.score(x_test, y_test)*100),2)
print ("\nModel Score:",lscore,"%")
```

# KNN

```python
from sklearn.neighbors import KNeighborsClassifier
kmodel = KNeighborsClassifier(n_neighbors=3)
kmodel.fit(x_train,y_train)
kpredictions = kmodel.predict(x_test)
print(kpredictions,y_test)

print("Confusion Matrix")
print(confusion_matrix(y_test,kpredictions))

kscore = round((kmodel.score(x_test, y_test)*100),2)
print ("\nModel Score:",kscore,"%")
```

```python
from sklearn.tree import DecisionTreeClassifier
dmodel = DecisionTreeClassifier()
dmodel.fit(x_train,y_train)
dprediction = dmodel.predict(x_test)
print(dprediction,y_test)

print("Confusion Matrix")
print(confusion_matrix(y_test,dprediction))

dscore = round((dmodel.score(x_test, y_test)*100),2)
print ("\nModel Score:",dscore,"%")
```

```python
from sklearn.svm import SVC
smodel = SVC()
smodel.fit(x_train,y_train)
sprediction = smodel.predict(x_test)
print(sprediction,y_test)

print("Confusion Matrix")
print(confusion_matrix(y_test,sprediction))

sscore = round((smodel.score(x_test, y_test)*100),2)
print ("\nModel Score:",sscore,"%")  
```

```python
data = [['Logistic Regression',lscore],['K-Nearest Neighbour',kscore],
        ['Decision Tree',dscore],['Support Vector Machine',sscore]]
final = pd.DataFrame(data,columns=['Algorithm','Precision'],index=[1,2,3,4])
```

```python
print("The results of Data Modeling are as follows:\n ")
print(final)  
```
